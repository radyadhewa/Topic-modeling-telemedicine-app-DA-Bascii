```python
from google_play_scraper import app, Sort, reviews
import pandas as pd
import numpy as np
```

# Halodoc Application

```python
result, continuation_token = reviews(
  'com.linkdokter.halodoc.android',
  lang='id',
  country='id',
  sort=Sort.NEWEST,
  count=2000,
  filter_score_with=None
)
```

```python
result
```

```python
df = pd.DataFrame(result)
df.head()
```

```python
df = df[['reviewId', 'content', 'score', 'thumbsUpCount', 'replyContent', 'appVersion']]
df.count()
```

```python
df.to_csv('halodoc_review_dataset.csv', index=False)
```
